Import potrzebnych bibliotek

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

Analiza produkcji kawy

```python
# Pobieramy caÅy dataset z produkcji
production = pd.read_csv('../coffee-datasets/Coffee_production.csv')
production
```

```python
# Pobieramy caÅy dataset z coffee yields czyli wydajnoÅÄ ziaren kawy mierzona w tonach na hektar
coffee_yields = pd.read_csv('../coffee-datasets/Coffee-yields.csv')
# My zajmujemy siÄ latami 1990-2020 wiÄc filtrujÄ Å¼eby tylko takie tu siÄ znajdowaÅy
coffee_yields = coffee_yields[(coffee_yields['Year'] > 1989) & (coffee_yields['Year'] < 2021)]
coffee_yields.rename(columns={coffee_yields.columns[3]: 'Tonnes per hectare'}, inplace=True)
coffee_yields
```

```python
# Pobieramy caÅy dataset z coffee production map dla wizualizacji na mapie
coffe_prod_world = pd.read_csv('../coffee-datasets/Coffe-bean-production-world.csv')
# My zajmujemy siÄ latami 1990-2020 wiÄc filtrujÄ Å¼eby tylko takie tu siÄ znajdowaÅy
coffe_prod_world = coffe_prod_world[(coffe_prod_world['Year'] > 1989) & (coffe_prod_world['Year'] < 2021)]
coffe_prod_world.rename(columns={coffe_prod_world.columns[3]: 'Tonnes'}, inplace=True)
coffe_prod_world
```

```python
# lista krajÃ³w i typÃ³w kaw dla ktÃ³rych robimy wizualizacjÄ danych
countries = production['Country']
coffe_types = production['Coffee type'].unique()

countries, coffe_types
```

```python
# CaÅkowita produkcja malejÄco (top 10)
top_production = production[['Country', 'Total_production']].sort_values(by=['Total_production'], ascending=False).head(10)

# Tutaj wyÅwietlam Total_production w zwykÅej notacji, nie eksponencjalnej
top_ten_producers = pd.concat([top_production['Country'], top_production['Total_production'].apply(lambda x: '{:,.0f}'.format(x))], axis=1).rename(columns={'Total_production': 'Total production'})
top_ten_producers
```

```python
# Pierwszy wykres - top 10 producentÃ³w kawy na Åwiecie
fig = px.bar(top_production, x='Country', y='Total_production', labels={
    "Country": "Countries",
    "Total_production": "Total Production (kg)"
}, title='Top 10 Countries with the Highest Coffee Production:', width=1200, height=800, color_discrete_sequence=['#78533d'])

fig.update_layout(
    yaxis=dict(title='Total Production (kg)'),
    annotations=[
        dict(
            text='(B = Billions)',
            showarrow=False,
            xref='paper',
            yref='y',
            x=1.0,
            y=max(top_production['Total_production']) * 0.95,
            font=dict(size=17)
        )
    ]
)

fig.show()
```

```python
# Drugi wykres - Årednia produkcja kawy na przeestrzeni lat 1990-2020
df = coffe_prod_world.copy()  # Skopiowanie oryginalnego DataFrame
df['Kilograms'] = df['Tonnes'] * 1000 # zamiana na kilogramy
df['Kilograms'] = df['Kilograms'].astype(float)  # Konwersja kolumny Tonnes per hectare na float

# Obliczanie Åredniej iloÅci ton na hektar dla kaÅ¼dego kraju
df_avg = df.groupby(['Entity', 'Code'])['Kilograms'].mean().reset_index()

# Tworzenie wizualizacji na mapie
fig = px.choropleth(df_avg, locations='Code', locationmode='ISO-3', color='Kilograms',
                    title='Average Coffee Production by Country (1990-2020)',
                    color_continuous_scale='Turbid',
                    hover_name='Entity',
                    hover_data=['Kilograms'],
                    labels={'Kilograms': 'Kilograms'},
                    height=800,
                    range_color=(0, df_avg['Kilograms'].max()))  # Zakres kolorÃ³w

fig.update_layout(geo=dict(showframe=False, showcoastlines=False, projection_type='equirectangular'))

# Ustawienie koloru i konturÃ³w dla krajÃ³w bez danych
fig.update_geos(
    visible=False,
    showcountries=True,
    countrycolor="white",
    showsubunits=True,
    subunitcolor="white",
    subunitwidth=0.5
)

# WyÅwietlanie wykresu
fig.show()
```

```python
# przygotowanie danych pod top 20 producentÃ³w kaw wzgledem typu kawy
top_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'], ascending=False).head(20)
top_production
```

```python
# trzeci wykres - Top 20 producentÃ³w kawy na Åwiecie wzgledem typu kawy
fig = px.bar(top_production, x='Country', y='Total_production', color='Coffee type', labels={
    "Country": "Countries",
    "Total_production": "Total Production"
},title='Top 20 Countries with Highest Coffee Production according to Coffee type', width=1000, height=600)

fig.update_layout(
    xaxis_title="Countries",
    yaxis_title="Total Production",
    title="Top 20 Countries with Highest Coffee Production according to Coffee type",
    width=1200,
    height=800,
    barmode='stack',
    yaxis=dict(title='Total Production (kg)'),
    annotations=[
        dict(
            text='(B = Billions)',
            showarrow=False,
            xref='paper',
            yref='y',
            x=1.0,
            y=max(top_production['Total_production']) * 0.95,
            font=dict(size=17)
        )
    ]
)
fig.show()
```

```python
# porozdzielamy sobie teraz kaÅ¼dy typ kawy
sorted_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'], ascending=False)
robusta = sorted_production[sorted_production['Coffee type'] == 'Robusta']
arabica = sorted_production[sorted_production['Coffee type'] == 'Arabica']
arabica_robusta = sorted_production[sorted_production['Coffee type'] == 'Arabica/Robusta']
robusta_arabica = sorted_production[sorted_production['Coffee type'] == 'Robusta/Arabica']
```

```python
# Czwarty wykres - Produkcja kawy podzielona na coffee type - wykres Treemap
fig = go.Figure()

# Definiowanie palety kolorÃ³w
colors = ["#78533d", "#8a6347", "#9e7352", "#b1845e", "#c5976b", "#d7aa79", "#e9be88", "#f4d2a0", "#f8e8c1", "#fdf3e3"]


# Dodawanie treemap dla Robusty
fig.add_trace(
    go.Treemap(
        labels=robusta.head(10).Country,
        parents=["Robusta"] * 10,
        values=robusta.Total_production,
        name="Robusta",
        text=robusta.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0, 0.45], y=[0.5, 1]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Arabiki
fig.add_trace(
    go.Treemap(
        labels=arabica.head(10).Country,
        parents=["Arabica"] * 10,
        values=arabica.Total_production,
        name="Arabica",
        text=arabica.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0.55, 1], y=[0.5, 1]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Arabiki/Robusty
fig.add_trace(
    go.Treemap(
        labels=arabica_robusta.head(10).Country,
        parents=["Arabica/Robusta"] * 10,
        values=arabica_robusta.Total_production,
        name="Arabica/Robusta",
        text=arabica_robusta.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0, 0.45], y=[0, 0.4]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Robusty/Arabiki
fig.add_trace(
    go.Treemap(
        labels=robusta_arabica.head(10).Country,
        parents=["Robusta/Arabica"] * 10,
        values=robusta_arabica.Total_production,
        name="Robusta/Arabica",
        text=robusta_arabica.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0.55, 1], y=[0, 0.4]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Aktualizowanie ukÅadu i wyÅwietlanie zdjÄcia
fig.update_layout(height=1200, width=1200, title_text="Coffee Production by Type")
fig.show()
```

```python
# Tutaj przygotowuje dfa pod wykres w ktÃ³rym pokaÅ¼emy produkcje w top 5 krajach na przestrzeni lat 1990-2020
top_production_helper = production[['Country', 'Total_production']].sort_values(by=['Total_production'], ascending=False).head(5)
top_production_by_years = production[production.Country.isin(top_production_helper['Country'].unique())]
top_production_by_years = top_production_by_years.T
top_production_by_years.drop(['Country', 'Total_production', 'Coffee type'], axis=0, inplace=True)
top_production_by_years.index.name = 'Year'
top_production_by_years.columns = top_production_helper['Country'].values
top_production_by_years
```

```python
# PiÄty wykres - top 5 krajÃ³w produkcja na przestrzeni lat 1990-2020
fig = px.line(top_production_by_years,
              x=top_production_by_years.index, y=top_production_by_years.columns[0:],
              labels={"Year": "Year", "value": "Production Rate"},
              title='Production Rates of the top 5 Countries over the years 1990-2020',
              width=1200, height=800)

fig.update_layout(legend_title_text="Country",
                  annotations=[dict(text='(B = Billions)',
                                    showarrow=False,
                                    xref='paper',
                                    yref='y',
                                    x=1.0,
                                    y=max(top_production['Total_production']) * 0.06,
                                    font=dict(size=17))])

fig.show()
```

```python
# SzÃ³sty wykres - wydajnoÅÄ kawy - iloÅÄ wyprodukowanej kawy na hektar pola - Årednia z lat 1990-2020
df = coffee_yields.copy()  # Skopiowanie oryginalnego DataFrame
df['Kilograms per hectare'] = df['Tonnes per hectare'] * 1000
df['Kilograms per hectare'] = df['Kilograms per hectare'].astype(float)  # Konwersja kolumny Tonnes per hectare na float

# Obliczanie Åredniej iloÅci ton na hektar dla kaÅ¼dego kraju
df_avg = df.groupby(['Entity', 'Code'])['Kilograms per hectare'].mean().reset_index()

# Tworzenie wizualizacji na mapie
fig = px.choropleth(df_avg, locations='Code', locationmode='ISO-3', color='Kilograms per hectare',
                    title='Average Coffee Yields by Country (1990-2020)',
                    color_continuous_scale='Turbid',
                    hover_name='Entity',
                    hover_data=['Kilograms per hectare'],
                    labels={'Kilograms per hectare': 'Kilograms per hectare'},
                    height=800,
                    range_color=(0, df_avg['Kilograms per hectare'].max()))  # Zakres kolorÃ³w

fig.update_layout(geo=dict(showframe=False, showcoastlines=False, projection_type='equirectangular'))

# Ustawienie koloru i konturÃ³w dla krajÃ³w bez danych
fig.update_geos(
    visible=False,
    showcountries=True,
    countrycolor="white",
    showsubunits=True,
    subunitcolor="white",
    subunitwidth=0.5
)

# WyÅwietlanie wykresu
fig.show()
```
