Importujemy wszystkie potrzebne biblioteki
Ogólnie w kolumnie z Bolivią usunąłem (Plurinational State of) bo wadziło w plotach

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

Analiza produkcji kawy

```python
# Pobieramy cały dataset z produkcji
production = pd.read_csv('coffee-dataset/Coffee_production.csv')
production
```

```python
# lista krajów i typów kaw dla których robimy wizualizację danych
countries = production['Country']
coffe_types = production['Coffee type'].unique()

countries, coffe_types
```

```python
# Całkowita produkcja malejąco
top_production = production[['Country', 'Total_production']].sort_values(by=['Total_production'], ascending=False).head(10)

# Tutaj wyświetlam Total_production w zwykłej notacji, nie eksponencjalnej (Tylko top 10 producentów)
top_ten_producers = pd.concat([top_production['Country'], top_production['Total_production'].apply(lambda x: '{:,.0f}'.format(x))], axis=1).rename(columns={'Total_production': 'Total production'})
top_ten_producers
```

```python
# Pierwszy plot - top 10 producentów kawy na świecie
fig = px.bar(top_production, x='Country', y='Total_production', labels={
    "Country": "Countries",
    "Total_production": "Total Production (kg)"
}, title='Top 10 Countries with the Highest Coffee Production:', width=1200, height=800, color_discrete_sequence=['#78533d'])

fig.update_layout(
    yaxis=dict(title='Total Production (kg)'),
    annotations=[
        dict(
            text='(B = Billions)',
            showarrow=False,
            xref='paper',
            yref='y',
            x=1.0,
            y=max(top_production['Total_production']) * 0.95,
            font=dict(size=17)
        )
    ]
)

fig.show()

```

```python
# przygotowanie danych pod top 20 producentów kaw wzgledem typu kawy
top_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'], ascending=False)
top_production = top_production.head(20)
top_production
```

```python
# Drugi plot - top 20 producentów kawy na świecie wzgledem typu kawy - pamiętajcie, że Arabica/Robusta to znaczy że Arabica tam dominuje
fig = px.bar(top_production, x='Country', y='Total_production', color='Coffee type', labels={
    "Country": "Countries",
    "Total_production": "Total Production"
},title='Top 20 Countries with Highest Coffee Production according to Coffee type', width=1000, height=600)

fig.update_layout(
    xaxis_title="Countries",
    yaxis_title="Total Production",
    title="Top 20 Countries with Highest Coffee Production according to Coffee type",
    width=1200,
    height=800,
    barmode='stack',
    yaxis=dict(title='Total Production (kg)'),
    annotations=[
        dict(
            text='(B = Billions)',
            showarrow=False,
            xref='paper',
            yref='y',
            x=1.0,
            y=max(top_production['Total_production']) * 0.95,
            font=dict(size=17)
        )
    ]
)
fig.show()
```

```python
# porozdzielamy sobie teraz każdy typ kawy i zrobimy ploty z tego
sorted_production = production[['Country', 'Total_production', 'Coffee type']].sort_values(by=['Total_production'], ascending=False)
robusta = sorted_production[sorted_production['Coffee type'] == 'Robusta']
arabica = sorted_production[sorted_production['Coffee type'] == 'Arabica']
arabica_robusta = sorted_production[sorted_production['Coffee type'] == 'Arabica/Robusta']
robusta_arabica = sorted_production[sorted_production['Coffee type'] == 'Robusta/Arabica']
```

```python
# Tworzenie subplotów
fig = go.Figure()

# Definiowanie palety kolorów
colors = ["#78533d", "#8a6347", "#9e7352", "#b1845e", "#c5976b", "#d7aa79", "#e9be88", "#f4d2a0", "#f8e8c1", "#fdf3e3"]


# Dodawanie treemap dla Robusty
fig.add_trace(
    go.Treemap(
        labels=robusta.head(10).Country,
        parents=["Robusta"] * 10,
        values=robusta.Total_production,
        name="Robusta",
        text=robusta.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0, 0.45], y=[0.5, 1]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Arabiki
fig.add_trace(
    go.Treemap(
        labels=arabica.head(10).Country,
        parents=["Arabica"] * 10,
        values=arabica.Total_production,
        name="Arabica",
        text=arabica.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0.55, 1], y=[0.5, 1]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Arabiki/Robusty
fig.add_trace(
    go.Treemap(
        labels=arabica_robusta.head(10).Country,
        parents=["Arabica/Robusta"] * 10,
        values=arabica_robusta.Total_production,
        name="Arabica/Robusta",
        text=arabica_robusta.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0, 0.45], y=[0, 0.4]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Dodawanie treemap dla Robusty/Arabiki
fig.add_trace(
    go.Treemap(
        labels=robusta_arabica.head(10).Country,
        parents=["Robusta/Arabica"] * 10,
        values=robusta_arabica.Total_production,
        name="Robusta/Arabica",
        text=robusta_arabica.Total_production.apply(lambda x: f"{x:,.2f}"),
        textinfo="text+label",
        domain=dict(x=[0.55, 1], y=[0, 0.4]),
        marker=dict(colors=colors),
        hovertemplate='<b>%{label}</b><br>Total Production: %{value:,.2f}<br>'
    )
)

# Aktualizowanie układu i wyświetlanie zdjęcia
fig.update_layout(height=1200, width=1200, title_text="Coffee Production by Type")
fig.show()
```

```python
# Tutaj przygotowuje dfa pod wykres w którym pokażemy produkcje w top 5 krajach na przestrzeni lat 1990-2020
top_production_helper = production[['Country', 'Total_production']].sort_values(by=['Total_production'], ascending=False).head(5)
top_production_by_years = production[production.Country.isin(top_production_helper['Country'].unique())]
top_production_by_years = top_production_by_years.T
top_production_by_years.drop(['Country', 'Total_production', 'Coffee type'], axis=0, inplace=True)
top_production_by_years.index.name = 'Year'
top_production_by_years.columns = top_production_helper['Country'].values
top_production_by_years
```

```python
# Czwarty plot = top 5 krajów produkcja na przestrzeni lat 1990-2020
fig = px.line(top_production_by_years,
              x=top_production_by_years.index, y=top_production_by_years.columns[0:],
              labels={"Year": "Year", "value": "Production Rate"},
              title='Production Rates of the top 5 Countries over the years 1990-2020',
              width=1200, height=800)

fig.update_layout(legend_title_text="Country",
                  annotations=[dict(text='(B = Billions)',
                                    showarrow=False,
                                    xref='paper',
                                    yref='y',
                                    x=1.0,
                                    y=max(top_production['Total_production']) * 0.06,
                                    font=dict(size=17))])

fig.show()
```
