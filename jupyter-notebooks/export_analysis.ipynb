```python
import pandas as pd
import plotly.express as px
```

```python
# Pobieramy dane eksportu kawy
coffee_export = pd.read_csv('coffee-dataset/Coffee_export.csv')
coffee_export
```

```python
# Top 10 eksporterów kawy na świecie i ich total eksport na przestrzeni 30 lat 1990-2020
top_export = coffee_export[['Country', 'Total_export']].sort_values(by=['Total_export'], ascending=False).head(10)
top_export
```

```python
# Pierwszy wykres - Top 10 coffee exporters
# Definiowanie palety barw
colors = ["#78533d", "#8a6347", "#9e7352", "#b1845e", "#c5976b", "#d7aa79", "#e9be88", "#f4d2a0", "#f8e8c1", "#fdf3e3"]

fig = px.pie(top_export, values='Total_export', names='Country', title='Top Coffee Exporters')

fig.update_traces(textposition='inside', textinfo='percent+label', marker=dict(colors=colors))

fig.update_layout(width=1000, height=1000)

fig.show()
```

```python
# Eksport kawy top 5 eksporterów na przestrzeni lat
top_5_export = coffee_export[['Country', 'Total_export']].sort_values(by=['Total_export'], ascending=False).head(5)
top_exportation = coffee_export[coffee_export.Country.isin(top_5_export['Country'].unique())]
top_exportation = top_exportation.T
top_exportation.drop(['Country', 'Total_export'], axis=0, inplace=True)
top_exportation.index.name = 'Year'
top_exportation.columns = top_5_export['Country'].values
top_exportation['Brazil'] = abs(top_exportation['Brazil'])
top_exportation
```

```python
# Eksport kawy top 5 eksporterów na przestrzeni lat
fig = px.line(top_exportation, x = top_exportation.index, y= top_exportation.columns[0:],labels={
    "Country": "Countries",
    "value": "Export"}, title='Countrywise Export Rate of Coffee',width=1200, height=800)
fig.show()
```
